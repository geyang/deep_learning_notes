```python
def σ(z):
    return 1/(1 + e**(-z))
```

```python
def Plot(fn, *args):
    argLength = len(args);
    if argLength == 1:
        start = args[0][0]
        end = args[0][1]
        points = None
        try: 
            points = args[0][2]
        except:
            pass
        if not points: points = 30
        xs = linspace(start, end, points);
        plot(xs, list(map(fn, xs)));
```

```python
Plot(σ, [-10, 10])
```
